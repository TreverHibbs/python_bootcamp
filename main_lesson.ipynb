# CSV to XLS
## A python workshop

Start by importing pandas and converting the csv file to xlsx

```python
import pandas as pd

read_file = pd.read_csv (r'./grades.csv')
read_file.to_excel (r'./converted_file.xlsx', index = None, header=True)
```

Next, read the xlsx file. This will create a pandas ExcelFile object.

```python
data = pd.ExcelFile('converted_file.xlsx')
```

We can use the data object to edit the information in the xlsx file. First lets print out the data so we can see what we are working with.

```python
print(data.parse())
```
