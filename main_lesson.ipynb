An index is a dataFrame attribute that we can use to select very specific pieces of information from any number of columns. 

In the following example we will use the `contains()` method of the `str` attribute of the pandas index to select only the rows that that have publish_dates in October. 

```python
import re
df = df[df['publish_date'].str.contains('10/[0-9]/2017','10/1[1-2]/2017', regex = True)]
```
